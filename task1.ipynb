```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing dataset

data = pd.read_csv("eCommerce.csv")
```

```python
# printing first 5 rows

data.head()
```

```python
# names of all columns

data.columns
```

```python
# taking useful columns into another dataset

data1 = data[['Customer ID', 'Age', 'Gender', ' Total Spending ', 'Website Visits']]
```

```python
# printing first 5 rows

data1.head()
```

```python
data1.shape
```

```python
# Create a histogram for ages

plt.figure(figsize=(10, 5))  # Set the size of the figure
plt.hist(data1['Age'], bins=50, color='lightpink', edgecolor='red')  # 'Age' is the column name
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.grid(False)
plt.show()
```

```python
# Create a bar chart for genders

gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(3,3))
plt.bar(gender_counts.index, gender_counts.values, color='lightgreen', alpha=0.7)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders')
plt.xticks(rotation=45)
# plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
